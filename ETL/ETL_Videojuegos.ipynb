Vamos a realizar ETL. Para lo cual comenzaré importando las librerias que necesito

```python
import pandas as pd  
import json 
import gzip
import ast
```

Crearé una función para ver el archivo 'stem_games.json.gz' que está comprimido y luego crearé un dataset df_games

```python
def ver():
    with gzip.open('../DATA/steam_games.json.gz','rt', encoding='utf-8') as f:
        return [json.loads(line.strip())
            for line in f]

data_games = ver()

df_games = pd.DataFrame(data_games)
```

Eliminaré las filas que contienen valores nulos desde la columna de posicion 0 a la de posicion 14.Porque luego eliminaré las ultimas 4 columnas porque se encuentran en los otros archivos. 

```python
df_games = df_games.dropna(subset=df_games.iloc[:, :15].columns, how='all')
```

```python
df_games.info()
```

Eliminar columnas porque estas columnas ya las tengo en otros archivos

```python
df_games= df_games.drop('user_id', axis=1)
```

```python
df_games=df_games.drop('steam_id', axis=1)
```

```python
df_games=df_games.drop('items', axis=1)
```

```python
df_games=df_games.drop('items_count', axis=1)
```

Voy a descomprimir el archivo user_reviews.json.gz y crearé un dataset df_reviews

Crearé una columna recommend en el dataset reviews

Realizaré el analisis de sentimiento

Vamos a cambiar el tipo de dato de la columna review1 a cadena de texto (string)

```python
df_reviews['review1'] = df_reviews['review1'].astype(str)
```

Vamos a cambiar a míniscula la columna review1 

```python
df_reviews['review1'] = df_reviews['review1'].str.lower()
```

Importaré la clase TextBlob del módulo textblob.TextBlob es una herramienta de procesamiento de lenguaje 
natural (NLP) que se utiliza para realizar tareas como análisis de sentimientos.

```python
from textblob import TextBlob
```

Aplicare el analisis de sentimiento a la columna review1 utilizando el metodo sentiment.polarity

```python
df_reviews['sentiment_analysis'] = df_reviews['review1'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
df_reviews['sentiment_analysis']
```

```python
# Asignare una escala de valor según el análisis de sentimiento y crearé la columna score
df_reviews['score'] = df_reviews['sentiment_analysis'].apply(lambda x: 0 if x < -0.1 else (1 if x < 0.1 else 2))
```

```python
df_reviews['score']
```

```python
df_reviews
```
