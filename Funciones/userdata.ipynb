porcentaje de recomendacion / cantidad de items

```python
df_reviews=pd.read_csv('../ETL/reviews.csv')
```

```python
#Cuenta las recomendaciones por usuario
recomendaciones_por_usuario = df_reviews.groupby('user_id')['recommend'].count().reset_index()
```

```python
df_items=pd.read_csv('../ETL/items.csv')
```

```python
df_games_original= pd.read_csv('../ETL/games_original.csv')
```

```python
df_items_count= df_games_original.groupby('user_id')['items_count'].sum().reset_index()
```

```python
df_items_count
```

```python
df_funcion1= pd.merge(recomendaciones_por_usuario, df_items_count, on='user_id')
```

```python
df_funcion1
```

Agrego una columna de porcentaje de recomendacion que divide a recommend en items_count y lo multiplica por cien

```python
df_funcion1['porcentaje de recomendación'] = (df_funcion1['recommend'] / df_funcion1['items_count']) * 100
#Redondeo en dos decimales
df_funcion1['porcentaje de recomendación'] = df_funcion1['porcentaje de recomendación'].round(2)
```

```python
df_funcion1
```

Eliminaré la columna recommend que era necesaria para crear el porcentaje de recomendacion. 

```python
df_funcion1=df_funcion1.drop('recommend', axis=1)
```

```python
df_funcion1
```

Ahora uniré los dataset dinero_gastado y df_funcion1 en un dataset userdata1

```python
userdata1= pd.merge(dinero_gastado, df_funcion1, on='user_id')
```

```python
userdata1
```

Cambiaré el orden de las columnas

```python
userdata1= userdata1.reindex(columns=['user_id','dinero_gastado','porcentaje de recomendación', 'items_count'])
```

```python
userdata1
```

Crearé un csv del dataset userdata1

```python
userdata1.to_csv('userdata1.csv',index= False, errors ='replace')
```

Crearé la funcion 

```python
def userdata(user_id):
    usuario = userdata1[userdata1['user_id'] == user_id]
    
    if len(usuario) == 0:
        return "El usuario no existe."
    
    dinero_gastado = usuario['dinero_gastado'].values[0]
    porcentaje_recomendacion = usuario['porcentaje de recomendación'].values[0]
    items_count = usuario['items_count'].values[0]
    
    return dinero_gastado, porcentaje_recomendacion, items_count

```

Ejemplo

```python
user_id = '--ace--'
dinero_gastado, porcentaje_recomendacion, items_count = userdata(user_id)

print(f"Dinero gastado por el usuario {user_id}: {dinero_gastado}")
print(f"Porcentaje de recomendación para el usuario {user_id}: {porcentaje_recomendacion}")
print(f"Cantidad de items para el usuario {user_id}: {items_count}")
```
